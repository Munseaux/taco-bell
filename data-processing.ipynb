```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=True)
```

```python
url = "https://tacobell.fandom.com/wiki/Category:Food_items"
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
menu_item = soup.find_all('a', class_="category-page__member-link")
results = []
for item in menu_item:
    results.append(item.text)

lowercase = []

for result in results:
    lowercase.append(result.lower())

sub = "category:"

for item in lowercase:
    if item.find(sub) != -1:
        lowercase.remove(item)
    

lowercase


```

```python
import csv

old_menu = []
with open('tacobellmenu.csv') as f:
    reader = csv.reader(f)
    old_menu = list(reader)
clean_old_menu = []
old_menu
for item in old_menu:
    clean_old_menu.append(item[0].lower())

clean_old_menu

for item in clean_old_menu:
    if item not in lowercase:
        lowercase.append(item)


df = pd.DataFrame(data={"menu item": lowercase})

df.to_csv('tacobellmenu.csv', index=False, header=False)
```
